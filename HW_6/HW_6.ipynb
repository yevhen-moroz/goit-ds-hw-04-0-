```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("Housing.csv")
df = df[["area", "bathrooms", "bedrooms", "price"]]


X = df[["area", "bathrooms", "bedrooms"]].values
y = df["price"].values.reshape(-1, 1)

scaler = StandardScaler()
X = scaler.fit_transform(X)
y_scaler = StandardScaler()
y = y_scaler.fit_transform(y)

X = np.hstack([np.ones((X.shape[0], 1)), X])
```

```python
def hypothesis(X, w):
    return np.dot(X, w)
```

```python
def compute_loss(X, y, w):
    m = len(y)
    predictions = hypothesis(X, w)
    errors = predictions - y
    return (1 / (2 * m)) * np.sum(errors ** 2)
```

```python
def gradient_step(X, y, w, lr):
    m = len(y)
    predictions = hypothesis(X, w)
    errors = predictions - y
    gradient = (1 / m) * np.dot(X.T, errors)
    return w - lr * gradient
```

```python
def train_gradient_descent(X, y, lr=0.001, epochs=1000):
    w = np.zeros((X.shape[1], 1))
    
    for i in range(epochs):
        w = gradient_step(X, y, w, lr)
        
        if i % 100 == 0:
            loss = compute_loss(X, y, w)
            print(f"Epoch {i}, Loss: {loss}")
    
    return w


w_gd = train_gradient_descent(X, y)

print("\nWeights (Gradient Descent):")
print(w_gd)
```

```python
def normal_equation(X, y):
    return np.dot(np.dot(np.linalg.inv(np.dot(X.T, X)), X.T), y)


w_ne = normal_equation(X, y)

print("\nWeights (Normal Equation):")
print(w_ne)
```

```python
model = LinearRegression()
model.fit(X[:, 1:], y)

print("\nSklearn:")
print("Intercept:", model.intercept_)
print("Coefficients:", model.coef_)
```
